```python
import pandas as pd
```

```python
data = pd.read_csv('1976-2020-president.csv')
```

```python
data
```

```python
data = data.loc[(data['party_simplified'] == 'DEMOCRAT') | (data['party_simplified'] == 'REPUBLICAN')]
data = data[["year", "state", "candidate","candidatevotes", "totalvotes","party_simplified"]]
data.to_csv('crap.csv')
```

```python
democrats = data.loc[(data['party_simplified'] == 'DEMOCRAT')]
democrats

republicans = data.loc[(data['party_simplified'] == 'REPUBLICAN')]
republicans
```

```python
data_filtered = pd.merge(democrats, republicans, how="left", on=["totalvotes", "state", "year"])
data_filtered = data_filtered.rename(columns={"candidate_x": "candidate_dem", 
                  "candidatevotes_x": "votes_dem", 
                  "party_simplified_x": "dem",
                  "candidate_y": "candidate_rep", 
                  "candidatevotes_y": "votes_rep", 
                  "party_simplified_y": "rep"})
data_filtered

data_filtered = data_filtered[['year', 'state', 'dem', 'rep', 'candidate_dem', 'candidate_rep', 'votes_dem', 'votes_rep', 'totalvotes']]
data_filtered
```

```python
data_filtered = data_filtered
data_filtered["percent_dem"] = ((data_filtered["votes_dem"]/data_filtered["totalvotes"])*100)
data_filtered["percent_rep"] = ((data_filtered["votes_rep"]/data_filtered["totalvotes"])*100)
```

```python
data_filtered.to_csv('data.csv')
```
